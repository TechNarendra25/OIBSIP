# 📥 Load the Iris Dataset

## Overview

The Iris dataset is one of the most well-known datasets in Machine Learning. It was introduced by **Ronald A. Fisher** in 1936 and is commonly used to demonstrate classification algorithms.

This dataset is available directly in the **Scikit-learn** library, so no external download is required.

### Dataset Information

- Total Samples: **150**
- Features: **4**
- Classes: **3**
- Target Variable: **Species**

### Features

- Sepal Length (cm)
- Sepal Width (cm)
- Petal Length (cm)
- Petal Width (cm)

### Target Classes

- Setosa
- Versicolor
- Virginica

```python
# ==========================================================
# Load the Iris Dataset
# ==========================================================

# Load dataset
iris = load_iris()

# Create DataFrame
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# Add target column
df["species"] = iris.target

# Convert numeric labels to species names
df["species"] = df["species"].map({
    0: "Setosa",
    1: "Versicolor",
    2: "Virginica"
})

print("✅ Dataset loaded successfully.")
```

# 📌 Feature Scaling

Feature scaling is a preprocessing technique used to bring numerical features onto a similar scale.

In this project, feature scaling is **not applied** because:

- The Iris dataset contains features with relatively similar ranges.
- Tree-based algorithms such as Decision Tree and Random Forest do not require feature scaling.
- The objective of this internship task is to compare multiple classification algorithms using the original dataset.

For larger or more complex datasets, feature scaling (e.g., StandardScaler or MinMaxScaler) is often recommended, especially for distance-based algorithms like K-Nearest Neighbors (KNN).

# ✂️ Train-Test Split

## Overview

Before training a Machine Learning model, the dataset is divided into two parts:

- **Training Set (80%)** – Used to train the model.
- **Testing Set (20%)** – Used to evaluate the model.

This ensures that the model is tested on unseen data, providing a fair estimate of its real-world performance.

### Why use `random_state = 42`?

Setting a fixed random state ensures that the train-test split remains the same every time the notebook is executed, making the results reproducible.

```python
# ==========================================================
# Train-Test Split
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("✅ Dataset split completed successfully.")
```

```python
print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)

print()

print("Training Target   :", y_train.shape)
print("Testing Target    :", y_test.shape)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
print("Training Set Distribution")

print(y_train.value_counts())

print()

print("Testing Set Distribution")

print(y_test.value_counts())
```

### 📌 Observation

The dataset has been successfully divided into training and testing sets using an 80:20 ratio.

The `stratify` parameter ensures that the proportion of each Iris species remains the same in both datasets. This prevents class imbalance and leads to a more reliable evaluation of the Machine Learning models.

# 🤖 Machine Learning Model Building

## Overview

Machine Learning models learn patterns from historical data and use those patterns to make predictions on unseen data.

In this project, we compare four classification algorithms:

1. Logistic Regression
2. K-Nearest Neighbors (KNN)
3. Decision Tree
4. Random Forest

Each model will be evaluated using:

- Accuracy Score
- Confusion Matrix
- Classification Report

Finally, the best-performing model will be selected based on its performance.

# 🔹 Model 1 : Logistic Regression

## Overview

Logistic Regression is a supervised Machine Learning algorithm used for classification problems.

It predicts the probability that a sample belongs to a particular class. Although its name contains "Regression", it is widely used for classification tasks.

```python
# ==========================================================
# Logistic Regression
# ==========================================================

lr = LogisticRegression(max_iter=200)

lr.fit(X_train, y_train)

lr_pred = lr.predict(X_test)

print("✅ Logistic Regression Model Trained Successfully")
```

```python
lr_accuracy = accuracy_score(y_test, lr_pred)

print(f"Accuracy : {lr_accuracy:.4f}")
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    confusion_matrix(y_test, lr_pred),
    annot=True,
    cmap="Blues",
    fmt="d"
)

plt.title("Logistic Regression Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("../images/confusion_matrix_lr.png", dpi=300)

plt.show()
```

```python
print(classification_report(y_test, lr_pred))
```

### 📌 Observation

Logistic Regression achieved a high classification accuracy on the Iris dataset.

Most flower species were classified correctly with only a few minor misclassifications.

# 🔹 Model 2 : K-Nearest Neighbors (KNN)

## Overview

K-Nearest Neighbors (KNN) is a distance-based Machine Learning algorithm.

It classifies a new sample by identifying the nearest training samples and assigning the majority class among them.

```python
# ==========================================================
# KNN
# ==========================================================

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

knn_pred = knn.predict(X_test)

print("✅ KNN Model Trained Successfully")
```

```python
knn_accuracy = accuracy_score(y_test, knn_pred)

print(f"Accuracy : {knn_accuracy:.4f}")
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    confusion_matrix(y_test, knn_pred),
    annot=True,
    cmap="Greens",
    fmt="d"
)

plt.title("KNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("../images/confusion_matrix_knn.png", dpi=300)

plt.show()
```

```python
print(classification_report(y_test, knn_pred))
```

### 📌 Observation

KNN performs exceptionally well because the Iris dataset contains well-separated classes, especially for the Setosa species.

# 🔹 Model 3 : Decision Tree

## Overview

Decision Tree is a supervised learning algorithm that makes decisions using a tree-like structure of rules.

It is easy to interpret and works well for both classification and regression tasks.

```python
dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)

print("✅ Decision Tree Model Trained Successfully")
```

```python
dt_accuracy = accuracy_score(y_test, dt_pred)

print(f"Accuracy : {dt_accuracy:.4f}")
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    confusion_matrix(y_test, dt_pred),
    annot=True,
    cmap="Oranges",
    fmt="d"
)

plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("../images/confusion_matrix_dt.png", dpi=300)

plt.show()
```

```python
print(classification_report(y_test, dt_pred))
```

### 📌 Observation

Decision Tree performs well on the Iris dataset by learning simple decision rules based on flower measurements.

# 🔹 Model 4 : Random Forest

## Overview

Random Forest is an ensemble learning algorithm that combines multiple Decision Trees to improve prediction accuracy and reduce overfitting.

It is one of the most widely used classification algorithms in Machine Learning.

```python
rf = RandomForestClassifier(
    random_state=42
)

rf.fit(X_train, y_train)

rf_pred = rf.predict(X_test)

print("✅ Random Forest Model Trained Successfully")
```

```python
rf_accuracy = accuracy_score(y_test, rf_pred)

print(f"Accuracy : {rf_accuracy:.4f}")
```

```python
plt.figure(figsize=(6,5))

sns.heatmap(
    confusion_matrix(y_test, rf_pred),
    annot=True,
    cmap="Purples",
    fmt="d"
)

plt.title("Random Forest Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.savefig("../images/confusion_matrix_rf.png", dpi=300)

plt.show()
```

```python
print(classification_report(y_test, rf_pred))
```

### 📌 Observation

Random Forest generally provides excellent performance because it combines multiple decision trees, making the model more robust and reducing overfitting.

# 📊 Model Comparison

## Overview

After training multiple Machine Learning models, it is important to compare their performance.

The comparison is based on the **Accuracy Score**, which measures the percentage of correctly classified samples.

The model with the highest accuracy and consistent performance will be selected as the best model for this project.

```python
# ==========================================================
# Model Comparison
# ==========================================================

comparison = pd.DataFrame({
    "Model": [
        "Logistic Regression",
        "K-Nearest Neighbors",
        "Decision Tree",
        "Random Forest"
    ],
    "Accuracy": [
        lr_accuracy,
        knn_accuracy,
        dt_accuracy,
        rf_accuracy
    ]
})

comparison = comparison.sort_values(
    by="Accuracy",
    ascending=False
).reset_index(drop=True)

comparison
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=comparison,
    x="Accuracy",
    y="Model",
    palette="viridis"
)

plt.title("Model Accuracy Comparison")

plt.xlim(0.90,1.01)

plt.savefig("../images/model_comparison.png", dpi=300)

plt.show()
```

### 📌 Observation

All four Machine Learning models performed exceptionally well on the Iris dataset.

Among them, the model with the highest accuracy is selected as the final model for prediction.

# 🏆 Best Model Selection

The best-performing model is selected based on:

- Highest Accuracy
- Better Generalization
- Consistent Classification Performance

```python
best_model = comparison.iloc[0]

print("🏆 Best Model")
print("-"*40)

print("Model    :", best_model["Model"])
print("Accuracy :", round(best_model["Accuracy"]*100,2),"%")
```

### 📌 Observation
